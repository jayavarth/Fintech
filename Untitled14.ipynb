```python
import pandas as pd

users_df = pd.read_excel("/content/users_sheet_50_rows.xlsx")
expenses_df = pd.read_excel("/content/expenses_sheet_50_rows.xlsx")
investments_df = pd.read_excel("/content/investment_recommendations_50_rows.xlsx")
```

```python
user_investment_df = users_df.merge(investments_df, on="Risk Level", how="left")
```

```python
expense_summary = expenses_df.groupby("UserID")["Amount"].sum().reset_index()
expense_summary.rename(columns={"Amount": "Total Expenses"}, inplace=True)
```

```python
final_data = user_investment_df.merge(expense_summary, on="UserID", how="left")
final_data["Total Expenses"] = final_data["Total Expenses"].fillna(0)
final_data["Remaining Balance"] = final_data["Income"] - final_data["Total Expenses"]
```

```python
def recommend_budgeting(user_id):
    try:
        user = final_data[final_data["UserID"] == user_id]
        if user.empty:
            print(f"User ID {user_id} not found!\n" + "-" * 50)
            return
        user = user.iloc[0]
        print(f"User:{user['Name']}")
        print(f"Income:{user['Income']}")
        print(f"Total Expenses:{user['Total Expenses']}")
        print(f"Savings:{user['Savings']}")
        print(f"Remaining Balance:{user['Remaining Balance']}")
        user_expenses = expenses_df[expenses_df["UserID"] == user_id]
        if not user_expenses.empty:
            highest_expense_category = user_expenses.groupby("Category")["Amount"].sum().idxmax()
            highest_expense_amount = user_expenses.groupby("Category")["Amount"].sum().max()
            print(f"Major Spending Area: {highest_expense_category} (${highest_expense_amount})")
        if user['Savings'] < 0:
            print(f"Suggestion: Consider reducing expenses in {highest_expense_category} to improve savings.")
        elif user['Remaining Balance'] <= 0:
            print(f"Warning: You are ov
            erspending! Try cutting down on {highest_expense_category}.")
        else:
            print(f"Suggested Investment: {user['Investment Option']} "
                  f"with an expected return of {user['Expected Return (%)']}")
        print("-" * 10)
    except Exception as e:
        print(f"Error processing UserID {user_id}: {e}")
for user_id in users_df["UserID"]:
    recommend_budgeting(user_id)
```
